```python
# Create point geometries
geometry = geopandas.points_from_xy(df2.longitud, df2.latitud)
df2 = geopandas.GeoDataFrame(df2, geometry=geometry)
df2 = df2.set_crs(epsg=4326, allow_override=True)

df2
```

```python
map = folium.Map(location=[df2.latitud.mean(), df2.longitud.mean()], 
                 zoom_start=6, 
                 control_scale=True)
```

```python
#Loop through each row in the dataframe
for i,row in df2.iterrows():
    #Setup the content of the popup
    iframe = folium.IFrame('nombre:' + str(row["nombre"]))
    
    #Initialise the popup using the iframe
    popup = folium.Popup(iframe, min_width=300, max_width=300)
    
    #Add each row to the map
    folium.Marker(location=[row['latitud'],row['longitud']],
                  popup = popup, c=row['institucion']).add_to(map)
```

#### Correlation based on stations

```python
df13
```

```python
df14 = df13.groupby('variable')[['value','ppt_mm']].corr().unstack().iloc[:,1]
df14 = pd.DataFrame(df14.reset_index())
df14
```

```python
df15 = df13.groupby('variable')[['value','ppt_mm']].count().reset_index()
df15
```

```python
df16 = df14.merge(df15,left_on='variable', right_on='variable')
df16 = df16.iloc[:, 1:4]
df16.columns = ['codigo_estacion', 'pearson', 'count']
df16
```

```python
df16
```

```python
df13['variable'] = df13['variable'].astype('str')
df16['codigo_estacion'] = df16['codigo_estacion'].astype('str')
df17 = df13[['variable', 'nombre', 'COD_CUEN_x','COD_SUBC_x','COD_SSUBC_x','NOM_SSUBC_x','POINT_X','POINT_Y']].groupby(["nombre",'variable','COD_CUEN_x','COD_SUBC_x','COD_SSUBC_x','NOM_SSUBC_x']).mean().reset_index()

df18 = df16.merge(df17,left_on='codigo_estacion',right_on='variable',how='left')
df18
```

```python
df18['r2'] = df18['pearson']** 2
df18 = df18.sort_values(by=['r2'], ascending=False)
df18 = df18.dropna(subset=['pearson']).reset_index(drop=True)
df18
```

```python
import pandas as pd
import folium

# df18.to_excel('results/precipitation_record_and_estimation_final.xlsx', index=False)
df18 = pd.read_excel('results/precipitation_record_and_estimation_final.xlsx')
```

```python
# Create a map object centered on the mean of the coordinates
mean_lat = df18['POINT_Y'].mean()
mean_lon = df18['POINT_X'].mean()
world_map = folium.Map(location=[mean_lat, mean_lon], zoom_start=4)

# Add markers for each location in the DataFrame
for i in range(0, len(df18)):
    if df18.iloc[i]['r2'] >= 0.6:
        color = 'green'
    elif df18.iloc[i]['r2'] < 0.6:
        color = 'red'
    else:
        color = 'black'
    folium.Marker(location=[df18.iloc[i]['POINT_Y'], df18.iloc[i]['POINT_X']],
                  popup=f"{df18.iloc[i]['nombre']} ({df18.iloc[i]['codigo_estacion']}) ({df18.iloc[i]['r2']}) ({df18.iloc[i]['count']})",
                  icon=folium.Icon(color=color)).add_to(world_map)


# Display the map
world_map
```
